```python
from collections import Counter

items0 = [{"bags": Counter(), "value": 0} for _ in range(8)]
items1 = [{"bags": Counter(), "value": 0} for _ in range(8)]
items0[0]["value"] = 1
items1[5]["value"] = 1

xg = 9

def deep_copy_items(items):
    return [{"bags": Counter(item["bags"]), "value": item["value"]} for item in items]


def is_valid(items0, items1):
    return not any(
        c > 1 or k == 0 for item in items0 + items1 for k, c in item["bags"].items()
    )

base0 = deep_copy_items(items0)
base1 = deep_copy_items(items1)

for i in range(1, xg):
    a0 = deep_copy_items(base0)
    a1 = deep_copy_items(base1)

    a0[1]["value"] = a1[4]["value"] = i
    a0[0]["bags"][abs(a0[0]["value"] - a0[1]["value"])] += 1
    a0[1]["bags"][abs(a0[0]["value"] - a0[1]["value"])] += 1
    a1[5]["bags"][abs(a1[5]["value"] - a1[4]["value"])] += 1
    a1[4]["bags"][abs(a1[5]["value"] - a1[4]["value"])] += 1

    if not is_valid(a0, a1):
        continue

    for j in range(1, xg):
        b0 = deep_copy_items(a0)
        b1 = deep_copy_items(a1)

        b0[2]["value"] = b1[6]["value"] = j
        b0[0]["bags"][abs(b0[0]["value"] - b0[2]["value"])] += 1
        b0[2]["bags"][abs(b0[0]["value"] - b0[2]["value"])] += 1
        b1[5]["bags"][abs(b1[5]["value"] - b1[6]["value"])] += 1
        b1[6]["bags"][abs(b1[5]["value"] - b1[6]["value"])] += 1

        if not is_valid(b0, b1):
            continue

        for k in range(1, xg):
            c0 = deep_copy_items(b0)
            c1 = deep_copy_items(b1)

            c0[3]["value"] = c1[7]["value"] = k
            c0[3]["bags"][abs(c0[3]["value"] - c0[2]["value"])] += 1
            c0[3]["bags"][abs(c0[3]["value"] - c0[1]["value"])] += 1
            c0[2]["bags"][abs(c0[3]["value"] - c0[2]["value"])] += 1
            c0[1]["bags"][abs(c0[3]["value"] - c0[1]["value"])] += 1
            c1[7]["bags"][abs(c1[7]["value"] - c1[6]["value"])] += 1
            c1[7]["bags"][abs(c1[7]["value"] - c1[4]["value"])] += 1
            c1[4]["bags"][abs(c1[7]["value"] - c1[4]["value"])] += 1
            c1[6]["bags"][abs(c1[7]["value"] - c1[6]["value"])] += 1

            if not is_valid(c0, c1):
                continue

            for l in range(1, xg):
                d0 = deep_copy_items(c0)
                d1 = deep_copy_items(c1)

                d0[4]["value"] = d1[1]["value"] = l
                d0[1]["bags"][abs(d0[1]["value"] - d1[1]["value"])] += 1
                d1[1]["bags"][abs(d0[1]["value"] - d1[1]["value"])] += 1
                d0[4]["bags"][abs(d0[4]["value"] - d1[4]["value"])] += 1
                d1[4]["bags"][abs(d0[4]["value"] - d1[4]["value"])] += 1
                d0[4]["bags"][abs(d0[4]["value"] - d0[2]["value"])] += 1
                d0[2]["bags"][abs(d0[4]["value"] - d0[2]["value"])] += 1
                d1[1]["bags"][abs(d1[1]["value"] - d1[6]["value"])] += 1
                d1[6]["bags"][abs(d1[1]["value"] - d1[6]["value"])] += 1

                if not is_valid(d0, d1):
                    continue

                for m in range(1, xg):
                    e0 = deep_copy_items(d0)
                    e1 = deep_copy_items(d1)

                    e0[5]["value"] = e1[0]["value"] = m
                    e0[5]["bags"][abs(e0[5]["value"] - e1[5]["value"])] += 1
                    e1[5]["bags"][abs(e0[5]["value"] - e1[5]["value"])] += 1
                    e0[0]["bags"][abs(e0[0]["value"] - e1[0]["value"])] += 1
                    e1[0]["bags"][abs(e0[0]["value"] - e1[0]["value"])] += 1
                    e0[5]["bags"][abs(e0[5]["value"] - e0[3]["value"])] += 1
                    e0[5]["bags"][abs(e0[5]["value"] - e0[4]["value"])] += 1
                    e0[3]["bags"][abs(e0[5]["value"] - e0[3]["value"])] += 1
                    e0[4]["bags"][abs(e0[5]["value"] - e0[4]["value"])] += 1
                    e1[0]["bags"][abs(e1[0]["value"] - e1[1]["value"])] += 1
                    e1[0]["bags"][abs(e1[0]["value"] - e1[7]["value"])] += 1
                    e1[1]["bags"][abs(e1[0]["value"] - e1[1]["value"])] += 1
                    e1[7]["bags"][abs(e1[0]["value"] - e1[7]["value"])] += 1

                    if not is_valid(e0, e1):
                        continue

                    for n in range(1, xg):
                        f0 = deep_copy_items(e0)
                        f1 = deep_copy_items(e1)

                        f0[6]["value"] = f1[3]["value"] = n
                        f0[6]["bags"][abs(f0[6]["value"] - f1[6]["value"])] += 1
                        f1[6]["bags"][abs(f0[6]["value"] - f1[6]["value"])] += 1
                        f0[3]["bags"][abs(f0[3]["value"] - f1[3]["value"])] += 1
                        f1[3]["bags"][abs(f0[3]["value"] - f1[3]["value"])] += 1
                        f0[6]["bags"][abs(f0[6]["value"] - f0[1]["value"])] += 1
                        f0[6]["bags"][abs(f0[6]["value"] - f0[5]["value"])] += 1
                        f0[5]["bags"][abs(f0[6]["value"] - f0[1]["value"])] += 1
                        f0[1]["bags"][abs(f0[6]["value"] - f0[5]["value"])] += 1
                        f1[3]["bags"][abs(f1[3]["value"] - f1[1]["value"])] += 1
                        f1[3]["bags"][abs(f1[3]["value"] - f1[5]["value"])] += 1
                        f1[1]["bags"][abs(f1[3]["value"] - f1[1]["value"])] += 1
                        f1[5]["bags"][abs(f1[3]["value"] - f1[5]["value"])] += 1

                        if not is_valid(f0, f1):
                            continue

                        for o in range(1, xg):
                            g0 = deep_copy_items(f0)
                            g1 = deep_copy_items(f1)

                            g0[7]["value"] = g1[2]["value"] = o
                            g0[2]["bags"][abs(g0[2]["value"] - g1[2]["value"])] += 1
                            g1[2]["bags"][abs(g0[2]["value"] - g1[2]["value"])] += 1
                            g0[7]["bags"][abs(g0[7]["value"] - g1[7]["value"])] += 1
                            g1[7]["bags"][abs(g0[7]["value"] - g1[7]["value"])] += 1
                            g0[7]["bags"][abs(g0[7]["value"] - g0[4]["value"])] += 1
                            g0[7]["bags"][abs(g0[7]["value"] - g0[6]["value"])] += 1
                            g0[7]["bags"][abs(g0[7]["value"] - g0[0]["value"])] += 1
                            g0[4]["bags"][abs(g0[7]["value"] - g0[4]["value"])] += 1
                            g0[6]["bags"][abs(g0[7]["value"] - g0[6]["value"])] += 1
                            g0[0]["bags"][abs(g0[7]["value"] - g0[0]["value"])] += 1
                            g1[2]["bags"][abs(g1[2]["value"] - g1[3]["value"])] += 1
                            g1[2]["bags"][abs(g1[2]["value"] - g1[4]["value"])] += 1
                            g1[2]["bags"][abs(g1[2]["value"] - g1[0]["value"])] += 1
                            g1[3]["bags"][abs(g1[2]["value"] - g1[3]["value"])] += 1
                            g1[4]["bags"][abs(g1[2]["value"] - g1[4]["value"])] += 1
                            g1[0]["bags"][abs(g1[2]["value"] - g1[0]["value"])] += 1

                            if not is_valid(g0, g1):
                                continue

                            print("solution")
                            print(f"i={i}, j={j}, k={k}, l={l}, m={m}, n={n}, o={o}")
                            for item in g0:
                                print(item)
```
